```python
from diffusers import StableDiffusionXLPipeline, EulerAncestralDiscreteScheduler
import torch
```

# Creating Pipeline(model)

```python
model = "Linaqruf/animagine-xl-2.0"
negative_prompt = "lowres, bad anatomy, bad hands, text, error, missing fingers, extra digit, fewer digits, cropped, worst quality, low quality, normal quality, jpeg artifacts, signature, watermark, username, blurry"
```

```python
def Animage_generator(Prompt):
  pipe = StableDiffusionXLPipeline.from_pretrained(
      model,
      torch_dtype = torch.float16,
      use_safetensors = True,
      variant = "fp16",
  )
  pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
  pipe.to("cuda")

  pipe.interference_batch_size = 1

  image = pipe(
      Prompt,
      negative_prompt = negative_prompt,
      width = 512,
      height = 512,
      guidance_scale = 12,
      target_size = (512, 512),
      original_size = (4096, 4096),
      num_inference_steps = 50
  ).images[0]

  return image
```

# Generate Images and an Interface

```python
import gradio as gr
```

```python
iface = gr.Interface(fn = Animage_generator, inputs = "text" , outputs = "image", examples = [['Magical ambiance, lovable, unmatched artistry, best in class, 1teenager, silver hair, stylish jacket, confident stare, upper torso, rooftop setting, midnight, choker'],
 ['Radiant personality, endearing, masterpiece creation, exceptional quality, 1female character, orange hair, casual attire, friendly expression, waist-up, park scene, starry night, beret.']])

iface.launch(debug = True)
```
